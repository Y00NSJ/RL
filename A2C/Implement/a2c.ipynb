```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import gym
```

```python
from a2c_learn import A2Cagent
```

```python
# 환경 이름 지정
env_name = 'Pendulum-v1'
env = gym.make(env_name)

# A2C 에이전트 생성
agent = A2Cagent(env)

# 학습 실행
max_episode_num = 1000
agent.train(max_episode_num)
```

```python
import numpy as np
import matplotlib.pyplot as plt

rewards = np.loadtxt('./save_weights/pendulum_epi_reward.txt')

plt.plot(rewards)
plt.title("Pendulum Episode Rewards")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.grid(True)
plt.show()
```

```python
env = gym.make('Pendulum-v1')
agent = A2Cagent(env)

dummy_input = tf.random.normal((1, agent.state_dim))
agent.actor(dummy_input)
agent.critic(dummy_input)

agent.load_weights('./save_weights/')

state, _ = env.reset()
time = 0

while True:
    env.render()

    # 액터 모델을 통해 행동 추론
    action = agent.actor(tf.convert_to_tensor([state], dtype=tf.float32))[0][0].numpy()

    state, reward, terminated, truncated, _ = env.step(action)
    done = terminated or truncated
    time += 1

    print('Time:', time, 'Reward:', reward)

    if done:
        break

env.close()
```
